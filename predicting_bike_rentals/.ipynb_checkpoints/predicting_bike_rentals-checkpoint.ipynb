# Predicting Bike Rentals - Error Comparison
Bike sharing has become more and more popular in big cities. Some days, the weather will be awesome, and the bike rentals will skyrocket. Some other days, the weather may be still be awesome, but for some reason the bike rentals are low -e.g. it's holiday and people go outside VS it's Wednesday and people are at work.

How can we predict the demand (number of bikes rented) based on measurable factors? In this project, we will build a prediction model that estimates the bikes rented per day. We will be using a CSV file that contains following data -thanks Hadi Fanaee-T for creating this data set!:

- `instant` - A unique sequential ID number for each row
- `dteday` - The date of the rentals
- `season` - The season in which the rentals occurred
- `yr` - The year the rentals occurred
- `mnth` - The month the rentals occurred
- `hr` - The hour the rentals occurred
- `holiday` - Whether or not the day was a holiday
- `weekday` - The day of the week (as a number, 0 to 7)
- `workingday` - Whether or not the day was a working day
- `weathersit` - The weather (as a categorical variable)
- `temp` - The temperature, on a 0-1 scale
- `atemp` - The adjusted temperature
- `hum` - The humidity, on a 0-1 scale
- `windspeed` - The wind speed, on a 0-1 scale
- `casual` - The number of casual riders (people who hadn't previously signed up with the bike sharing program)
- `registered` - The number of registered riders (people who had already signed up)
- `cnt` - The total number of bike rentals (casual + registered)

We will compare the error between following algorithms:
- Linear regression
- Decision trees
- Random forests

## Building our model
Error metric selection: MSE -it works well with continuous values

```python
train = hour.sample(frac=.8,random_state=1)
test = hour[~hour.index.isin(train.index)]
```

```python
len(train) + len(test) == len(hour) # double-checking creation of train/test
```

### Feature selection
Some of the features are derived from other features. In order to avoid overfitting, we will eliminate those.

```python
features = list(train.columns)
features.remove("cnt")
features.remove("casual")
features.remove("registered")
features.remove("dteday")
print(features)
```

### Linear regression error

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(train[features],train['cnt'])
predictions = lr.predict(test[features])

mse = mean_squared_error(predictions,test['cnt'])
mse
```

Incredibly high error with linear regression!

### Decision tree error

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(min_samples_leaf=5)
dt.fit(train[features],train['cnt'])
predictions = dt.predict(test[features])

mse = mean_squared_error(predictions,test['cnt'])
mse
```

### Random forest error

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(min_samples_leaf=5)
rf.fit(train[features],train['cnt'])
predictions = rf.predict(test[features])

mse = mean_squared_error(predictions,test['cnt'])
mse
```
